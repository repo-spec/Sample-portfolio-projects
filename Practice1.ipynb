## 
1. Develops and trains multiple polynomial regression models, with orders 2, 3, and 5, that use one attribute of a data frame as the source variable and another as a target variable.
2. Calculate and display the MSE and R^2 values for the trained models.
3. Compare the performance of the models.


```python
X=df[['RAM_GB']]
Y=df['Price']
```

```python
lr1=LinearRegression()
lr2=LinearRegression()
```

```python
poly=PolynomialFeatures(degree=2)
poly1=PolynomialFeatures(degree=3)
poly2=PolynomialFeatures(degree=5)

X_poly=poly.fit_transform(X)
X_poly1=poly1.fit_transform(X)
X_poly2=poly2.fit_transform(X)

lr.fit(X_poly,Y)
y_pred=lr.predict(X_poly)
print('R2 score of polynomial 1 is',r2_score(Y,y_pred))
print('MSE of polynomial 1 is ',mean_squared_error(Y,y_pred))

lr1.fit(X_poly1,Y)
y_pred=lr1.predict(X_poly1)
print('R2 score of polynomial 2 is ',r2_score(Y,y_pred))
print('MSE of polynomial 2 is ',mean_squared_error(Y,y_pred))



lr2.fit(X_poly2,Y)
y_pred=lr2.predict(X_poly2)
print('R2 score of polynomial 3 is ',r2_score(Y,y_pred))
print('MSE of polynomial 3 is ',mean_squared_error(Y,y_pred))

```

## 
The performance of the model for polynomials of different degree is almost same.  Polynomial of degree 5 performs slightly better than the others.


##
1. Create a pipeline that performs parameter scaling, Polynomial Feature generation, and Linear regression. Use the set of multiple features as before to create this pipeline.
2. Calculate and display the MSE and R^2 values for the trained model.


```python
X=df.drop(columns=['Price','Manufacturer','Price-binned'])
Y=df['Price']
```

```python
pipe=make_pipeline(StandardScaler(),PolynomialFeatures(degree=2),LinearRegression())
```

```python
pipe.fit(X,Y)
y_pred=pipe.predict(X)

print('R2 score is',r2_score(Y,y_pred))
print('MSE is ',mean_squared_error(Y,y_pred))
```

## 
1. Use polynomial features for some of the attributes of a data frame.
2. Perform Grid search on a ridge regression model for a set of values of hyperparameter alpha and polynomial features as input.
3. Use cross-validation in the Grid search.
4. Evaluate the resulting model's MSE and R^2 values.


```python
poly=PolynomialFeatures()
X_poly=poly.fit_transform(X)
param_grid={'alpha':[0.1,1,10]}
model=Ridge()
grid_search=GridSearchCV(model,param_grid,cv=5)
grid_search.fit(X_poly,Y)
y_pred=grid_search.predict(X_poly)

print('R2 score is ',r2_score(Y,y_pred))
print('MSE is ',mean_squared_error(Y,y_pred))
```
